```python
df1= pd.read_csv('NFLX_stock_price.csv')
df2= pd.read_csv('DIS_stock_price.csv')
```

```python
#graficos de high price de cada plataforma 

disney = df1['High Price']
netflix = df2['High Price']
plt.hist(disney)
plt.hist(netflix)
plt.xlabel('platafromas')
plt.ylabel('PRECIOS')
plt.title('High Price de cada platafromar')
plt.show()
```
